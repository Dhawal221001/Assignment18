## Q1. What is Flask Framework? What are the advantages of Flask Framework?

### Flask is a web development framework developed in Python.Flask is known as a micro-framework because it is lightweight and only provides components that are essential. It only provides the necessary components for web development, such as routing, request handling, sessions.

## Advantages of flask:
### It is a lightweight framework that offers hassle-free development
### Provide flexibility to the developer to experiment with their modules or architecture
### Offers a built-in development server and fast debugger
### Easily scalable for the applications
### Support for secure cookies

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

### url_for() function is used in flask for URL building
